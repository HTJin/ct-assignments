# Sum of Pairs
#### https://www.codewars.com/kata/54d81488b981293527000c8f

```python
# dictionary solution, pretty much took what we got from class but got rid of the target variable declaration
def sum_pairs_1(ints, s):
    my_dict = {}
    for num in ints:
        if s - num in my_dict:
            return [s - num, num]
        my_dict[num] = num
    # return None # <-- I realize I don't need this line, but that's what I have submitted

# set solution, all the cool kids in the solutions page were using it
def sum_pairs_2(ints, s):
    my_set = set()
    for num in ints:
        if s - num in my_set:
            return [s - num, num]
        my_set.add(num)
```

```python
import codewars_test as test

@test.describe("Sum of Pairs Test")
def sample_tests():
    l1 = [1, 4, 8, 7, 3, 15]
    l2 = [1, -2, 3, 0, -6, 1]
    l3 = [20, -13, 40]
    l4 = [1, 2, 3, 4, 1, 0]
    l5 = [10, 5, 2, 3, 7, 5]
    l6 = [4, -2, 3, 3, 4]
    l7 = [0, 2, 0]
    l8 = [5, 9, 13, -3]
    l9 = [1] * 10000000
    l9[len(l9) // 2 - 1] = 6
    l9[len(l9) // 2] = 7
    l9[len(l9) - 2] = 8
    l9[len(l9) - 1] = -3
    l9[0] = 13
    l9[1] = 3

    @test.it("Dictionary Solution")
    def dictionary_solution():
        test.assert_equals(sum_pairs_1(l1, 8), [1, 7], "Basic: %s should return [1, 7] for sum = 8" % l1)
        test.assert_equals(sum_pairs_1(l2, -6), [0, -6], "Negatives: %s should return [0, -6] for sum = -6" % l2)
        test.expect(sum_pairs_1(l3, -7) is None, "No Match: %s should return None for sum = -7" % l3)
        test.assert_equals(sum_pairs_1(l4, 2), [1, 1], "First Match From Left: %s should return [1, 1] for sum = 2 " % l4)
        test.assert_equals(sum_pairs_1(l5, 10), [3, 7], "First Match From Left REDUX!: %s should return [3, 7] for sum = 10 " % l5)
        test.assert_equals(sum_pairs_1(l6, 8), [4, 4], "Duplicates: %s should return [4, 4] for sum = 8" % l6)
        test.assert_equals(sum_pairs_1(l7, 0), [0, 0], "Zeroes: %s should return [0, 0] for sum = 0" % l7)
        test.assert_equals(sum_pairs_1(l8, 10), [13, -3], "Subtraction: %s should return [13, -3] for sum = 10" % l8)
        test.assert_equals(sum_pairs_1(l9, 13), [6, 7], "Ten Million Numbers With Middle Pair: Should terminate with a valid pair output")
        test.assert_equals(sum_pairs_1(l9, 5), [8, -3], "Ten Million Numbers With Pair At End: Should terminate with a valid pair output")
        test.assert_equals(sum_pairs_1(l9, 16), [13, 3], "Ten Million Numbers With Pair At Start: Should terminate with a valid pair output")
        test.expect(sum_pairs_1(l9, 31) is None, "Ten Million Numbers With No Match: Should return None in a decent time period")
        
    @test.it("Set Solution")
    def set_solution():
        test.assert_equals(sum_pairs_2(l1, 8), [1, 7], "Basic: %s should return [1, 7] for sum = 8" % l1)
        test.assert_equals(sum_pairs_2(l2, -6), [0, -6], "Negatives: %s should return [0, -6] for sum = -6" % l2)
        test.expect(sum_pairs_2(l3, -7) is None, "No Match: %s should return None for sum = -7" % l3)
        test.assert_equals(sum_pairs_2(l4, 2), [1, 1], "First Match From Left: %s should return [1, 1] for sum = 2 " % l4)
        test.assert_equals(sum_pairs_2(l5, 10), [3, 7], "First Match From Left REDUX!: %s should return [3, 7] for sum = 10 " % l5)
        test.assert_equals(sum_pairs_2(l6, 8), [4, 4], "Duplicates: %s should return [4, 4] for sum = 8" % l6)
        test.assert_equals(sum_pairs_2(l7, 0), [0, 0], "Zeroes: %s should return [0, 0] for sum = 0" % l7)
        test.assert_equals(sum_pairs_2(l8, 10), [13, -3], "Subtraction: %s should return [13, -3] for sum = 10" % l8)
        test.assert_equals(sum_pairs_2(l9, 13), [6, 7], "Ten Million Numbers With Middle Pair: Should terminate with a valid pair output")
        test.assert_equals(sum_pairs_2(l9, 5), [8, -3], "Ten Million Numbers With Pair At End: Should terminate with a valid pair output")
        test.assert_equals(sum_pairs_2(l9, 16), [13, 3], "Ten Million Numbers With Pair At Start: Should terminate with a valid pair output")
        test.expect(sum_pairs_2(l9, 31) is None, "Ten Million Numbers With No Match: Should return None in a decent time period")
```
